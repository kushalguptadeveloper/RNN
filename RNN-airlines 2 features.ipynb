```python
from pandas import read_csv 
import numpy as np
```

```python
df = read_csv('international-airline-passengers.csv')
def f(str1):
    no = str1[-2:]
    return no
df["1mor"] = df.Month.apply(f)
df.drop("Month",axis = 1, inplace = True)
df
```

```python
values = df.values.astype('Float32')
values.shape
```

```python
train_size = int((values.shape[0] * 0.67))
test_size = values.shape[0]  - train_size
train = values[0:train_size]
test = values[train_size:]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range =(0,1))
train = scaler.fit_transform(train)
test = scaler.transform(test)
```

```python
def create_dataset(data, k):
    dataX, dataY = [],[]
    for i in range(data.shape[0] - k):
        x = []
        for j in range(i,i+k):
            l = [data[j,0],data[j,1]]
            x.append(l)
        #x = data[i:i + k, 1]
        y = data[i + k, 1]
        dataX.append(x)
        dataY.append(y)
    return np.array(dataX), np.array(dataY)
```

```python
look_back = 12
trainX, trainY = create_dataset(train, look_back)
testX, testY = create_dataset(test, look_back)
```

```python
trainX.shape
```

```python
trainX = np.reshape(trainX, (trainX.shape[0], trainX.shape[1], 2))
testX = np.reshape(testX, (testX.shape[0], testX.shape[1], 2))
trainX.shape
```

```python
from keras.models import Sequential
from keras.layers import SimpleRNN, Dense
```

```python
model = Sequential()
model.add(SimpleRNN(4, input_shape=(look_back, 2)))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(trainX, trainY, epochs=10, batch_size = 1)
model.summary()
```

```python
score = model.evaluate(testX,testY)
score
```
